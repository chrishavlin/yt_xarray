```python
import xarray as xr
import yt_xarray
import yt
import os
from cartopy.feature import NaturalEarthFeature
```

```python
ds = xr.open_dataset("/home/chavlin/hdd/data/yt_data/yt_sample_sets/cmip6/co23D_Emon_GISS-E2-1-G_piControl_r101i1p1f1_gn_185001-190012.nc")
```

```python
yt_ds = ds.yt.load_uniform_grid(fields=['co23D'], sel_dict={'time':4})
```

```python
slc = yt.SlicePlot(yt_ds, "altitude", ("stream", "co23D"))
slc.set_log(("stream", "co23D"), False)
slc._setup_plots()

bounds = NaturalEarthFeature(category='cultural', scale='110m', facecolor='none',
                             name='admin_0_countries')
slc[("stream", "co23D")].axes.add_feature(bounds, edgecolor='gray')
slc.show()
```
